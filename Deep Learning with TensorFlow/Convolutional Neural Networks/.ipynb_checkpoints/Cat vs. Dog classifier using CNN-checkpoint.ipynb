```python
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
# lets vew single images of both classes
cat8 = cv2.imread('CATS_DOGS/train/CAT/8.jpg')
dog8 = cv2.imread('CATS_DOGS/train/DOG/8.jpg')
```

```python
# convert to RGB since we used cv2 to read the files
cat8 = cv2.cvtColor(cat8,cv2.COLOR_BGR2RGB)
dog8 = cv2.cvtColor(dog8,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(cat8)
```

```python
plt.imshow(dog8)
```

```python
print(cat8.shape)
print(dog8.shape)
#notice how the size is different
```

```python
'''
For our model to be more robust, it should be invariant to scale, lighting, rotation etc.
ImageDataGenerator will allow us to generate more images by including these variations
'''
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
image_gen = ImageDataGenerator(rotation_range = 30, 
                               width_shift_range = 0.12,
                              height_shift_range = 0.12,
                              rescale = 1/255,
                              shear_range = 0.2,
                              zoom_range = 0.25,
                              horizontal_flip = True,
                              fill_mode = 'nearest')
```

```python
plt.imshow(image_gen.random_transform(dog8))
```

```python
image_gen.flow_from_directory('CATS_DOGS/train')
```

```python
input_shape = (150,150,3)     #this is just a rough average of all input images
```

```python
from keras.models import Sequential
from keras.layers import Activation,Dropout,Flatten,Conv2D,MaxPooling2D,Dense
```

```python
model = Sequential()

model.add(Conv2D(filters = 32, kernel_size = (3,3), input_shape = input_shape, activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2,2)))

model.add(Conv2D(filters = 64, kernel_size = (3,3), input_shape = input_shape, activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2,2)))

model.add(Flatten())

model.add(Dense(128, activation = 'relu'))

model.add(Dropout(rate = 0.5))

model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
batch_size = 16
training_images = image_gen.flow_from_directory('CATS_DOGS/train',
                                                target_size = input_shape[:2],
                                                batch_size = batch_size,
                                                class_mode = 'binary')

# note: training_images already distinguished the images into X_train and Y_train 
```

```python
test_images = image_gen.flow_from_directory('CATS_DOGS/test',
                                                target_size = input_shape[:2],
                                                batch_size = batch_size,
                                                class_mode = 'binary')
```

```python
# we can see that the class labels have already been defined
training_images.class_indices
```

```python
# note: the model may not be able to read some images and throw warnings, we can ignore these using:
import warnings
warnings.filterwarnings('ignore')
```

```python
# fittng the model usign a single epoch causing training could take a lot of time
results = model.fit_generator(training_images, epochs = 1,
                              steps_per_epoch = 250, validation_data = test_images,
                             validation_steps = 12)   #A single epoch will be 250 batches of 16
```

```python
results.history['acc']
```
