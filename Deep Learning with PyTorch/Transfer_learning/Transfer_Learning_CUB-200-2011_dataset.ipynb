```python
import shutil

# save paths and create labels file
data_dir = 'CUB_200_2011'
img_dir = 'CUB_200_2011/images'
img_file = 'CUB_200_2011/images.txt'
labels_file = 'CUB_200_2011/image_class_labels.txt'

# creates a .csv file in img, label format for the CUB-200-2011 dataset
def create_labels_file(images_file, class_labels_file):

  path_to_file = 'CUB_200_2011/image_labels.csv'

  # read class labels
  labels_df = pd.read_csv(labels_file, delim_whitespace=True, index_col=False, header = None)
  labels_ser = labels_df[1]

  # read image names
  img_df = pd.read_csv(img_file, sep = '/', index_col=False, header = None)
  img_ser = img_df[1]

  # create new df formatted as image_name, class_label
  df = pd.concat([img_ser, labels_ser], axis = 1)
  df.to_csv(path_to_file, index = False, header = False)

  return path_to_file

def extract_images_from_folders(img_dir, out_dir):
  if not os.path.exists(out_dir):
    os.makedirs(out_dir)

    for root, dirs, files in os.walk(img_dir):
      for file in files:
        if file.endswith(('jpg','.jpeg','.png')):
          src_file_path = os.path.join(root, file)
          dst_file_path = os.path.join(out_dir, file)
          shutil.copy2(src_file_path, dst_file_path)
  print('Copying Complete')

# create annotations file and copy all images in a single folder
annotations_file = create_labels_file(img_file, labels_file)
final_img_dir = 'CUB_200_2011/all_images'
extract_images_from_folders(img_dir, final_img_dir)
```

```python
from PIL import Image

# custom dataset class
class CustomImageDataset(Dataset):
    def __init__(self, annotations_file, img_dir, transform=None):
        self.img_labels = pd.read_csv(annotations_file)
        self.img_dir = img_dir
        self.transform = transform

    def __len__(self):
        return len(self.img_labels)

    def __getitem__(self, idx):
        img_path = os.path.join(self.img_dir, self.img_labels.iloc[idx, 0])
        image = Image.open(img_path)
        label = self.img_labels.iloc[idx, 1]
        if self.transform:
            image = self.transform(image)
        return image, label
```

### Tranfer Learning using VGG16

```python
# define parameters
mean = [0.485, 0.456, 0.406]
std = [0.229, 0.224, 0.225]

# define transformer object
transform = transforms.Compose([transforms.Resize((224, 224)),
                                transforms.ToTensor(),
                              transforms.Normalize(mean, std)])

# create custom dataset object
dataset = CustomImageDataset(annotations_file, final_img_dir, transform=transform)
```

### Train Test Split

```python
from torch.utils.data import random_split
```

```python
train_ratio = 0.7
test_ratio = 1 - train_ratio

# get number of train and test samples
num_samp = len(dataset)
train_size = int(train_ratio * num_samp)
test_size = num_samp - train_size

train_data, test_data = random_split(dataset, [train_size, test_size])

# create dataloader objects
train_loader = DataLoader(train_data, batch_size=64, shuffle=True)
test_loader = DataLoader(test_data, batch_size=64, shuffle=False)
```

### Define Functions for Viewing Normalied Images

```python
def denormalize(tensor):
    tensor = tensor*std + mean
    return tensor

def show_img(img):
    # arrange channels
    img = img.numpy().transpose((1,2,0))

    # use mean and std values
    img = denormalize(img)

    # clip values and view image
    img = np.clip(img,0,1)
    plt.imshow(img)
```
