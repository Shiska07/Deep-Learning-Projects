```python
import re

# functions that extract patientid and view
def getPatientIDNumber(s):
    if 'P_' in s:
        result = re.search(r'P_(\d+)', s).group(1)
    else:
        result = re.search(r'P(\d+)', s).group(1)
    return result

def getview(s):
    if 'CC' in s:
        return 'CC'
    else:
        return 'MLO'

def getTrainInfo(s):
    if 'Training' in s:
        return 1
    else:
        return 0

def getType(s):
    if 'Mass' in s:
        return 'Mass'
    else:
        return 'Calc'
```

```python
df_dicom_info['patient_id'] = df_dicom_info['PatientID'].apply(getPatientIDNumber)
df_dicom_info['image view'] = df_dicom_info['PatientID'].apply(getview)
df_dicom_info['train'] = df_dicom_info['PatientID'].apply(getTrainInfo)
df_dicom_info['type'] = df_dicom_info['PatientID'].apply(getType)
```

### Create train and test dfs containing pathology info

### Create two separate dfs for train and test paths from 'df_dicom_info'


```python
df_train_paths = df_dicom_info[df_dicom_info['train'] == 1].copy()
df_test_paths = df_dicom_info[df_dicom_info['train'] == 0].copy()
```

```python
# print number of subjects in training and test paths df
print(df_train_paths['patient_id'].nunique())
print(df_test_paths['patient_id'].nunique())
```

```python
# create key column for all dataframes
def getkey(ser):
    ser = (ser.astype(object)).tolist()
    return '_'.join(ser)
```

```python
df_train_paths['key'] = [getkey(df_train_paths[['patient_id','Laterality']].iloc[i]) for i in range(len(df_train_paths))]
df_test_paths['key'] = [getkey(df_test_paths[['patient_id','Laterality']].iloc[i]) for i in range(len(df_test_paths))]
df_all_tr['key'] = [getkey(df_all_tr[['patient_id','Laterality']].iloc[i]) for i in range(len(df_all_tr))]
df_all_tst['key'] = [getkey(df_all_tst[['patient_id','Laterality']].iloc[i]) for i in range(len(df_all_tst))]
```

### Chck if the number of unique keys are same in corresponding dfs

```python
print(df_all_tr['key'].nunique())
print(df_train_paths['key'].nunique())
print(df_test_paths['key'].nunique())
print(df_all_tst['key'].nunique())
```

```python
# drop duplicated rows in pathology dfs
df_all_tr.drop_duplicates(inplace=True)
df_all_tst.drop_duplicates(inplace=True)
```

```python
def getPathologyTrain(s):
    p = df_all_tr[df_all_tr['key'] == s]['pathology'].values[0]
    return p

def getPathologyTest(s):
    p = df_all_tst[df_all_tst['key'] == s]['pathology'].values[0]
    return p
```

```python
df_train_paths['pathology'] = df_train_paths['key'].apply(getPathologyTrain)
df_test_paths['pathology'] = df_test_paths['key'].apply(getPathologyTest)
```
