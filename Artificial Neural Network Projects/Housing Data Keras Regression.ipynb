```python
df = df.drop('date', axis = 1)
```

```python
#too many categories so it will be dropped
df = df.drop('zipcode', axis = 1)
```

```python
df.head()
```

```python
df['yr_renovated'].value_counts()
```

```python
def renov(year):
    if year > 0:
        return 1
    else:
        return 0

df['renovated'] = df['yr_renovated'].apply(renov)
```

```python
df = df.drop('yr_renovated', axis = 1)
```

```python
df.head()
```

```python
df['sqft_basement'].value_counts()

#we can leave this as it is as 0 corresponds to no basement
```

```python
df = df[df['price'] < cutoff_price]          #dataframe without outliers:This means our model is only good a predicting values lower than cutoff_price
X = df.drop('price', axis = 1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state = 100)
```

```python
'''
Note: We scale our data only after the test train split because while scaling we have to fit our data and we don't want
to use our test data to fit the model since this will mean the model has prior info on the test data
'''
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
#Since we have 19 features, it would make sense to have atlease 19 neurons on each layer
```

```python
#MODEL CREATION
# Note: we don't really know how many layers to add or no. of epochs to use, this is a hyperparameter for now

model = Sequential()

model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))

model.add(Dense(1))           #output layer for price

model.compile(optimizer = 'adam', loss = 'mse')
```

```python
#MODEL TRAINING

# Note: A validation step will also be added this time. Validation data is only used for testing purposes, it does not 
# affect the weights and biases of the model. WE will also break down our data into batches 

model.fit(x = X_train, y = y_train, validation_data = (X_test, y_test), batch_size = 128, epochs = 400, verbose = 0)
```

```python
loss_df = pd.DataFrame(model.history.history)
```

```python
# Note: if you see the val_loss start to increase after a point, you might be overfitting
loss_df.plot(figsize = (10,6))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
mean_squared_error(y_test,predictions)
```

```python
mean_absolute_error(y_test,predictions)

#this is about 20% of mean price
```

```python
explained_variance_score(y_test, predictions)
```

```python
plt.figure(figsize = (10,6))
plt.scatter(y_test, predictions)
plt.plot(y_test,y_test, 'r')
```
