```python
# we will do feature engineering on the following attributes: 'size', 'revenue' and 'job state'
import math

def get_avg_size(s):
    values = []
    words = s.split()
    for word in words:
        word = word.strip('+')
        if word.isdigit():
            values.append(int(word))
    return math.ceil(sum(values)/len(values))
```

## Data Preparation for Training 

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler()
```

```python
# train test split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_feat, y, test_size=0.30, random_state = 100)
```

```python
# we do the scaling post-split because we want to prevent data leakage from the test set
X_train = sc.fit_transform(X_train)
```

```python
X_test = sc.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
X_train.shape
```

```python
y_train.shape
```

## Create Model

```python
model = Sequential()

# we have 33 input attributes
model.add(Dense(33, activation = 'relu'))
model.add(Dropout(0.5))

model.add(Dense(24, activation = 'relu'))
model.add(Dropout(0.15))

model.add(Dense(16, activation = 'relu'))
model.add(Dropout(0.15))

# we have 7 target class
model.add(Dense(7, activation = 'softmax'))

model.compile(optimizer = 'adam', loss = 'categorical_crossentropy')
```

```python
model.fit(x = X_train, y = y_train, epochs = 400, batch_size = 128, validation_data = (X_test, y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
# we can use earlystopping to prevent overfitting
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()

# we have 33 input attributes
model.add(Dense(33, activation = 'relu'))
model.add(Dropout(0.5))

model.add(Dense(24, activation = 'relu'))
model.add(Dropout(0.15))

model.add(Dense(16, activation = 'relu'))
model.add(Dropout(0.15))

# we have 7 target class
model.add(Dense(7, activation = 'softmax'))

model.compile(optimizer = 'adam', loss = 'categorical_crossentropy')
```

```python
# initialize an instance of early stopping
early_stop = EarlyStopping(monitor = 'val_loss', patience = 50, verbose = 1, mode = 'min')
```

```python
# refit model but add callback info
model.fit(x = X_train, y = y_train, epochs = 400, batch_size = 128, validation_data = (X_test, y_test), callbacks = [early_stop])
```
